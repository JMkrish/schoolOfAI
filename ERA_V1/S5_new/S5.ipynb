```python
class_map = {
    "ZERO": 0,
    "ONE": 1,
    "TWO": 2,
    "THREE": 3,
    "FOUR": 4,
    "FIVE": 5,
    "SIX": 6,
    "SEVEN": 7,
    "EIGHT": 8,
    "NINE": 9,
}

show_summary = True
show_dataset_analyze = False
```

### Get Available Device

```python
from utils import get_device

# Check GPU availability
use_cuda, device = get_device()

print(f"Device:{device}")
```

### Prepare datasets & loaders

```python
from transform import get_mnist_transform
from datasets import get_minst_dataset

train_transform = get_mnist_transform("train")
test_transform = get_mnist_transform("test")

train_data = get_minst_dataset("train", train_transform)
test_data = get_minst_dataset("test", test_transform)
```

```python
from utils import get_data_loader

kwargs = {
    "batch_size": 512,
    "shuffle": True,
    "num_workers": 2,
    "pin_memory": True,
}

test_loader = get_data_loader(test_data, kwargs)
train_loader = get_data_loader(train_data, kwargs)
```

## Calculate Class Scale

```python
from visualize import print_class_scale, print_samples

show_dataset_analyze = True
if show_dataset_analyze:
    print_class_scale(train_loader, class_map)
    print_samples(train_loader)
```

### Load model and get summary

```python
# from model import Net, model_summary
from model import Net

model = Net().to(device)

if show_summary:
    model.model_summary(input_size=(1, 28, 28))
```

### Visualize Model

```python
model.model_visualize(device, train_loader, "model")
```

### Run Model Pipeline

```python
import torch.nn.functional as F


from backpropagation import TTPipeline
from backpropagation import get_sgd_optimizer
from utils import get_lr_scheduler

ttp = TTPipeline(model, device)

optimizer = get_sgd_optimizer(model, lr=0.01, momentum=0.9)
scheduler = get_lr_scheduler(optimizer, step_size=15, gamma=0.1)

criterion = F.cross_entropy  #  F.nll_loss
num_epochs = 20

for epoch in range(1, num_epochs + 1):
    print(f"Epoch {epoch}")
    ttp.train(train_loader, optimizer, criterion)
    ttp.test(test_loader, criterion)
    scheduler.step()

ttp.print_performance()
```
